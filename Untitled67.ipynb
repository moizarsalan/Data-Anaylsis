This cell imports the pandas library and loads the data from the specified CSV file into a DataFrame called df. It then displays the first few rows of the DataFrame to give you a preview of the data.




```python
import pandas as pd

# Load CSV
f = "/content/Marketing Team Data - Sheet1.csv"
df = pd.read_csv(f)
df.head()
```

This cell removes the two columns named "@dropdown" and "@dropdown.1" from the DataFrame df. These columns seem to contain missing values and are not relevant to the analysis.

```python
# Drop unnecessary columns
df = df.drop(columns=["@dropdown", "@dropdown.1"])
```

his cell cleans the "Amount Spent in INR" column by removing '$' and ',' characters and then converts the column to a float data type. It also renames the cleaned column to "Amount Spent".

```python
# Clean and convert amount spent
df["Amount Spent"] = df["Amount Spent in INR"].replace('[\$,]', '', regex=True).astype(float)
```

This cell converts the "Click-Through Rate (CTR in %)" and "Unique Click-Through Rate (Unique CTR in %)" columns to float data types and renames them to "CTR" and "uCTR" respectively.

```python
# Convert CTRs to float
df["CTR"] = df["Click-Through Rate (CTR in %)"].astype(float)
df["uCTR"] = df["Unique Click-Through Rate (Unique CTR in %)"].astype(float)
```

This cell calculates the total clicks for each campaign by grouping the DataFrame by "Campaign Name" and summing the "Clicks" column. It then creates a horizontal bar plot to visualize the total clicks per campaign, sorted in ascending order.

```python
import matplotlib.pyplot as plt

# Total clicks
clk = df.groupby("Campaign Name")["Clicks"].sum().sort_values()

# Bar plot
clk.plot(kind="barh", color="skyblue")
plt.title("Total Clicks by Campaign")
plt.xlabel("Clicks")
plt.tight_layout()
plt.show()
```

his cell calculates the average Click-Through Rate (CTR) for each campaign. It groups the DataFrame by "Campaign Name", calculates the mean of the "CTR" column, sorts the results, and then generates a horizontal bar plot to visualize the average CTR per campaign.

```python
# Average CTR by campaign
ctr = df.groupby("Campaign Name")["CTR"].mean().sort_values()

# Bar plot
ctr.plot(kind="barh", color="salmon")
plt.title("Average CTR (%) by Campaign")
plt.xlabel("CTR (%)")
plt.tight_layout()
plt.show()
```

his cell calculates the total "Amount Spent" and "Clicks" for each campaign. It then creates a scatter plot to show the relationship between the total amount spent and the total clicks for each campaign.

```python
# Sum clicks & spend
sum_df = df.groupby("Campaign Name")[["Amount Spent", "Clicks"]].sum()

# Scatter plot
plt.scatter(sum_df["Amount Spent"], sum_df["Clicks"], color="green")
plt.title("Spend vs Clicks")
plt.xlabel("Amount Spent")
plt.ylabel("Total Clicks")
plt.grid(True)
plt.tight_layout()
plt.show()
```

This cell creates a pivot table to calculate the total "Reach" for each age group and campaign. It uses "Age" as the index, "Campaign Name" as the columns, and the sum of "Reach" as the values. Missing values are filled with 0.

```python
import matplotlib.pyplot as plt

# Pivot table: total reach by Age and Campaign
age_df = df.pivot_table(
    index="Age",
    columns="Campaign Name",
    values="Reach",
    aggfunc="sum"
).fillna(0)
```

This cell reindexes the age_df DataFrame to ensure the age groups are in a specific order (18-24, 25-34, etc.) for better visualization.

```python
# Sort age groups if needed
age_order = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
age_df = age_df.reindex(age_order)
```

This cell creates a stacked bar chart from the age_df pivot table to visualize the reach by age group and campaign. Each bar represents an age group, and the segments within the bar represent the reach for each campaign within that age group. The chart includes a title, axis labels, a legend, and a grid for better readability.

```python
# Create a stacked bar chart
ax = age_df.plot(
    kind="bar",
    stacked=True,
    colormap="tab20",
    figsize=(12, 6)
)

plt.title("Reach by Age Group and Campaign")
plt.ylabel("Reach")
plt.xlabel("Age Group")
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", title="Campaigns")
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

This cell cleans the "Cost Per Click (CPC)" column by removing non-numeric characters and extracting the first numerical value if multiple are present. It then converts the cleaned column to a float data type.

```python
# Keep only valid CPC values
df["CPC"] = df["Cost Per Click (CPC)"].replace('[^\d\.]', '', regex=True)

# Some rows have multiple values combined (like '$0.46$0.31$0.47') — keep only the first number
df["CPC"] = df["CPC"].str.extract(r'(\d+\.\d+)', expand=False).astype(float)
```

This cell groups the DataFrame by "Campaign Name" and calculates the mean of "Clicks", "CTR", "CPC", and "Amount Spent" for each campaign. It then normalizes these values to a 0-1 scale and creates a heatmap to visualize the normalized performance metrics across different campaigns.

The heatmap generated in cell 2MYkchr7lyLO visualizes the normalized performance metrics for each campaign. Here's a breakdown of what it shows:

Campaigns: Each row represents a different campaign.
* Metrics: Each column represents a different performance metric: "Clicks", "CTR", "CPC", and "Amount Spent".
* Normalized Values: The color intensity in each cell indicates the normalized value of that metric for that campaign. The values have been scaled to a range between 0 and 1.
* For "Clicks" and "CTR", higher normalized values (darker colors) indicate better performance (more clicks and higher click-through rates).
* For "CPC" (Cost Per Click) and "Amount Spent", lower normalized values (lighter colors) indicate better performance (lower cost per click and less spent).
* Color Scale: The color bar on the right shows the mapping from normalized values to colors. Lighter colors represent values closer to 0, and darker colors represent values closer to 1.
* Annotations: The numbers within each cell are the actual normalized values, providing a precise measure of performance for each metric and campaign.
How to Interpret the Heatmap:

By looking at the color intensity and the normalized values, you can quickly compare the performance of different campaigns across various metrics. For example:

* A campaign with a dark color in the "Clicks" column had a high number of clicks relative to other campaigns.
* A campaign with a light color in the "CPC" column had a low cost per click relative to other campaigns.
* You can also look for patterns across metrics. For instance, a campaign with high clicks and a low CPC might be considered very efficient.
* This heatmap provides a concise visual summary of how each campaign performed across these key metrics, making it easier to identify high-performing campaigns and areas where campaigns might need improvement.

```python
import pandas as pd
import seaborn as sns


# Group by campaign
hm_df = df.groupby("Campaign Name")[["Clicks", "CTR", "CPC", "Amount Spent"]].mean()

# Normalize values
hm_df_norm = (hm_df - hm_df.min()) / (hm_df.max() - hm_df.min())

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(hm_df_norm, annot=True, cmap="YlOrBr", linewidths=0.5, cbar_kws={'label': 'Normalized Value'})
plt.title("Heatmap of Campaign Performance Metrics")
plt.tight_layout()
plt.show()
```

his cell calculates the mean metrics ("CTR", "Clicks", "CPC", and "Amount Spent") for each campaign and normalizes them to a 0-1 scale. It then inverts the normalized scores for "CPC" and "Amount Spent" so that lower values in the original data correspond to higher normalized scores (indicating better performance).

This cell calculates an overall performance "Score" for each campaign by averaging the normalized and inverted metric scores. It then sorts the campaigns based on this score in descending order to rank them from best to worst performance and displays the ranked campaigns with their scores.

```python
# Group and round metrics
metrics_df = df.groupby("Campaign Name")[["CTR", "Clicks", "CPC", "Amount Spent"]].mean().round(2)

# Show the table
print(metrics_df.reset_index())
```

```python
# Group mean metrics by campaign
rank_df = df.groupby("Campaign Name")[["CTR", "Clicks", "CPC", "Amount Spent"]].mean()

# Normalize values to 0–1 scale
rank_norm = (rank_df - rank_df.min()) / (rank_df.max() - rank_df.min())

# Invert CPC and Amount Spent — lower is better
rank_norm["CPC"] = 1 - rank_norm["CPC"]
rank_norm["Amount Spent"] = 1 - rank_norm["Amount Spent"]
```

This cell merges the performance scores with the mean "CPC" and "Amount Spent" for each campaign. It then calculates the median "CPC" and "Efficiency" (performance score) across all campaigns. Finally, it identifies and displays the campaigns that have below-median efficiency and above-median CPC, highlighting those that are performing poorly relative to their cost.

```python
# Average normalized scores for a final performance score
rank_norm["Score"] = rank_norm.mean(axis=1)

# Sort by best performance
ranked_campaigns = rank_norm.sort_values("Score", ascending=False)

# Show rankings
ranked_campaigns[["Score"]].round(3)
```

```python
# Re-merge performance scores with original metrics
campaign_data = df.groupby("Campaign Name")[["CPC", "Amount Spent"]].mean()
campaign_data["Efficiency"] = ranked_campaigns["Score"]

# Calculate medians
cpc_median = campaign_data["CPC"].median()
eff_median = campaign_data["Efficiency"].median()

# Filter below-median efficiency & above-median CPC
to_remove = campaign_data[
    (campaign_data["Efficiency"] < eff_median) &
    (campaign_data["CPC"] > cpc_median)
]

# Round for clarity
to_remove = to_remove[["Efficiency", "CPC", "Amount Spent"]].round(3)

# Display
for name, row in to_remove.iterrows():
    print(f"- {name}\n  Efficiency: {row['Efficiency']} | CPC: ${row['CPC']} | Spent: ${row['Amount Spent']}")
```
